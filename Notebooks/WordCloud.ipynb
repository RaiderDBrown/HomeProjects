```python
from gensim.parsing.preprocessing import remove_stopwords
import re

def word_count(text):
    no_stopwords = remove_stopwords(text)
    trim_txt = re.sub("[^\w\s]", "", no_stopwords)
    counts = dict()
    words = trim_txt.split()
    
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts

data = word_count("""I would like to prioritize K-2 reading efforts - e.g. additional 
                  teaching assistants to support smaller groups of learning - so we can 
                  get as many students reading on grade level as possible as soon as possible. 
                  Priority going to Social-Emotional Support
                  """)
data
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(width=800, height=400, max_words=200, background_color='white').generate_from_frequencies(data)

plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
