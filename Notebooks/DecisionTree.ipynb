```python
# -*- coding: utf-8 -*-
"""
Created on Wed Apr 15 21:23:35 2020
@author: Owner
"""
import random
from sklearn import tree
#x1=[random.random() for i in range(10)]
#x2=[random.random() for i in range(10)]
X=[[random.random() for i in range(2)] for j in range(10)]
Y=[]
for i in X:
    Y.append(int((i[0]-i[1])*4))
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
tree.plot_tree(clf.fit(X,Y))
#vars(clf)
```
